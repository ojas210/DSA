```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


a = pd.read_csv('H:\DOWNLOADS\cars_data.csv')
a
```

```python
null_values=print(a.isnull().sum())
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8,5))
total = float(len(a))
ax = sns.countplot(x="Buyer Gender", hue="New Car", data=a)
plt.title('Trend of Male & Female customers', fontsize=20)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.show()
```
