```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


a = pd.read_excel('QA4ques.xlsx')
a
```

# 1 Compare Churn w.r.t gender

```python
#Compare Churn w.r.t gender
a1=a.loc[a['gender'] =='Male']
a1
sns.set(style="whitegrid")
plt.figure(figsize=(6,3))
total = float(len(a1))
a1x = sns.countplot(x="gender", hue="Churn", data=a1)
plt.title('Churn among Male', fontsize=20)
for p in a1x.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    a1x.annotate(percentage, (x, y),ha='center')
plt.show()
a2=a.loc[a['gender'] =='Female']
sns.set(style="whitegrid")
plt.figure(figsize=(6,3))
total = float(len(a2))
a2x = sns.countplot(x="gender", hue="Churn", data=a2)
plt.title('Churn among Female', fontsize=20)
for p in a2x.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    a2x.annotate(percentage, (x, y),ha='center')
plt.show()
```
